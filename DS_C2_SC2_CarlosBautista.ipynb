a) Extraer información del archivo solicitado

```python
import pandas as pd
vacunas=pd.read_csv('country_vaccinations.csv')
```

b)Estructura y tipos de datos

```python
vacunas.info()
```

b1) Las columnas date no son del tipo datetime64, se modifica

```python
vacunas['date']=pd.to_datetime(vacunas['date'])
vacunas.info()
```

c)Determinar la cantidad de vacunas aplicadas de cada compañía

```python
vacunas[['vaccines','daily_vaccinations']].groupby('vaccines').sum()
```

d)Cantidad de vacunas aplicadas en todo el mundo

```python
print('Total de vacunas en el mundo:',vacunas['daily_vaccinations'].sum())
```

e)Cantidad de vacunas aplicadas en promedio por país

```python
vacunas[['daily_vaccinations','country']].groupby('country').mean()
```

f)Vacunas aplicadas el día 29/01/2021

```python
vac_day=vacunas[vacunas['date'].isin(['29/01/21'])]
print('El total de vacunas aplicadas en ese día son',vac_day['daily_vaccinations'].sum())

                          
```

g)Dataframe conDiferencias

```python
conDiferencias=vacunas.copy()
conDiferencias['diferencias']=conDiferencias['daily_vaccinations']-conDiferencias['daily_vaccinations_raw']
conDiferencias
```

h)periodo del registro

```python
vacunas['date'].max()-vacunas['date'].min()
```

i) Nuevo dataframe conCantidad

```python
conCantidad=vacunas.copy()
conCantidad['canVac']=conCantidad['vaccines'].str.split('/').str.len()
conCantidad
```

j) Dataframe antes20 con los registros obtenidos antes del 20/12/20

```python
antes20=vacunas.loc[vacunas['date']<'20/12/20']
antes20

```

k) Dataframe con todos los registros donde se haya utilizado la vacuna Pfizer

```python

Pfizer= vacunas[vacunas['vaccines'].str.contains('pfizer', case=False)]
Pfizer
```

l)Almacenar los dataframes generados

```python
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="Sheet1")  
    conCantidad.to_excel(writer, sheet_name="Sheet2")
    antes20.to_excel(writer, sheet_name="Sheet3")  
    Pfizer.to_excel(writer, sheet_name="Sheet4")
```
